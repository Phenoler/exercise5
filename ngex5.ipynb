AIM #2: Finding issues (empty, NAs, incorrect value, incorrect format, outliers, etc.) 
1. Find out how many missing values there are in the dataset
2. For the 'Age' column, find the best way to handle the missing values
    2.1. Use an appropriate plot to study the nature of the 'Age' column
    2.2. Figure out what is the best way to calculate the central tendency of the 'Age' column based on the above plot
    2.3. Using the most suitable central tendency measure, fill the missing values in the age column
3. Decide what is the best way to handle the missing values in the 'Cabin' columns
4. Similarly, decide what is the best way to handle the missing values in the 'Embarked' columns
5. Handle the incorrect data under the 'Survived' columns using appropriate measure
6. Handle the incorrectly formatted data under the 'Fare' column


```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('titanic.csv')
missing_values = df.isnull().sum()
print(missing_values)
total_missing = missing_values.sum()
print(f'Total missing values: {total_missing}')
plt.figure(figsize=(10, 6))
plt.hist(df['Age'].dropna(), bins=30, edgecolor='black', alpha=0.7)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('count')
plt.grid(axis='y', alpha=0.75)
plt.show()
age_median = df['Age'].median()
df['Age'] = df['Age'].fillna(age_median)


df['Cabin'] = df['Cabin'].fillna('Unknown')
most_common_embarked = df['Embarked'].mode()[0]
df['Embarked'] = df['Embarked'].fillna(most_common_embarked)
incorrect_survived = df[~df['Survived'].isin([0, 1])]
print("Incorrect 'Survived' values:\n", incorrect_survived)
df = df[df['Survived'].isin([0, 1])]
df['Fare'] = pd.to_numeric(df['Fare'], errors='coerce')
fare_median = df['Fare'].median()
df['Fare'] = df['Fare'].fillna(fare_median)
```

AIM #3: Grouping 
1. Find out the average fare grouped by Pclass
    1.1. Plot the above using a suitable plot
2. Find out the average fare grouped by Sex
    2.1. Plot the above using a suitable plot

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('titanic.csv')
df['Fare'] = pd.to_numeric(df['Fare'], errors='coerce')
df['Fare'] = df['Fare'].fillna(df['Fare'].mean())
average_fare_by_pclass = df.groupby('Pclass')['Fare'].mean().reset_index()
print("Average Fare by Pclass:")
print(average_fare_by_pclass)
plt.figure(figsize=(8, 5))
plt.bar(average_fare_by_pclass['Pclass'].astype(str), average_fare_by_pclass['Fare'], color='lightpink')
plt.title('Average Fare by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Average Fare')
plt.xticks(ticks=[0, 1, 2], labels=['1st Class', '2nd Class', '3rd Class'])
plt.grid(axis='y')
plt.show()

```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('titanic.csv')
df['Fare'] = pd.to_numeric(df['Fare'], errors='coerce')
df['Fare'] = df['Fare'].fillna(df['Fare'].mean())
average_fare_by_sex = df.groupby('Sex')['Fare'].mean().reset_index()
print("Average Fare by Sex:")
print(average_fare_by_sex)
plt.figure(figsize=(8, 5))
plt.bar(average_fare_by_sex['Sex'], average_fare_by_sex['Fare'], color=['lightpink', 'lightblue'])
plt.title('Average Fare by Gender')
plt.xlabel('Gender')
plt.ylabel('Average Fare')
plt.grid(axis='y')
plt.show()
```

1. Plot the distribution of 'Age' using a suitable plot

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('titanic.csv')

data['Age'] = pd.to_numeric(data['Age'], errors='coerce')
data['Fare'] = pd.to_numeric(data['Fare'], errors='coerce')

data = data.dropna(subset=['Age', 'Fare'])

plt.figure(figsize=(10, 6))
data['Age'].hist(bins=30, color='blue', alpha=0.7, edgecolor='black')
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid(False)
plt.show()

plt.figure(figsize=(10, 6))
data['Fare'].hist(bins=30, color='green', alpha=0.7, edgecolor='black')
plt.title('Distribution of Fare')
plt.xlabel('Fare')
plt.ylabel('Frequency')
plt.grid(False)
plt.show()

pclass_counts = data['Pclass'].value_counts()
plt.figure(figsize=(10, 6))
pclass_counts.plot(kind='bar', color='orange', edgecolor='black')
plt.title('Distribution of Pclass')
plt.xlabel('Passenger Class')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.grid(False)
plt.show()

survived_counts = data['Survived'].value_counts()
plt.figure(figsize=(10, 6))
survived_counts.plot(kind='bar', color='red', edgecolor='black')
plt.title('Distribution of Survival')
plt.xlabel('Survived (0 = No, 1 = Yes)')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.grid(False)
plt.show()

embarked_counts = data['Embarked'].value_counts()
plt.figure(figsize=(10, 6))
embarked_counts.plot(kind='bar', color='purple', edgecolor='black')
plt.title('Distribution of Embarked Ports')
plt.xlabel('Embarked')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.grid(False)
plt.show()

plt.figure(figsize=(10, 6))
data.boxplot(column='Fare', by='Survived', grid=False)
plt.title('Fare Distribution by Survival')
plt.suptitle('')
plt.xlabel('Survived (0 = No, 1 = Yes)')
plt.ylabel('Fare')
plt.show()

plt.figure(figsize=(10, 6))
data.boxplot(column='Fare', by='Pclass', grid=False)
plt.title('Fare Distribution by Pclass')
plt.suptitle('')
plt.xlabel('Passenger Class')
plt.ylabel('Fare')
plt.show()

plt.figure(figsize=(10, 6))
data.boxplot(column='Age', by='Survived', grid=False)
plt.title('Age Distribution by Survival')
plt.suptitle('')
plt.xlabel('Survived (0 = No, 1 = Yes)')
plt.ylabel('Age')
plt.show()

plt.figure(figsize=(10, 6))
data.boxplot(column='Age', by='Pclass', grid=False)
plt.title('Age Distribution by Pclass')
plt.suptitle('')
plt.xlabel('Passenger Class')
plt.ylabel('Age')
plt.show()

data['Family_Size'] = data['SibSp'] + data['Parch']
plt.figure(figsize=(10, 6))
data.boxplot(column='Family_Size', by='Survived', grid=False)
plt.title('Family Size Distribution by Survival')
plt.suptitle('')
plt.xlabel('Survived (0 = No, 1 = Yes)')
plt.ylabel('Family Size')
plt.show()

plt.figure(figsize=(10, 6))
data.boxplot(column='Family_Size', by='Pclass', grid=False)
plt.title('Family Size Distribution by Pclass')
plt.suptitle('')
plt.xlabel('Passenger Class')
plt.ylabel('Family Size')
plt.show()

plt.figure(figsize=(10, 6))
plt.scatter(data['Age'], data['Fare'], alpha=0.5)
plt.title('Scatter Plot of Age vs Fare')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.grid(True)
plt.show()
```

AIM #5: Correlation

1. Generate a correlation matrix for the entire dataset
2. Find correlation between 'Age' and 'Fare'
3. What other possible correlations can be found in the dataset?

```python
import matplotlib.pyplot as plt
import csv

data = []

with open('titanic.csv', newline='', encoding='utf-8') as csvfile:
    reader = csv.reader(csvfile)
    headers = next(reader) 
    for row in reader:
        data.append(row)

survived = []
pclass = []
age = []
fare = []

for row in data:
    try:
        survived.append(int(row[0]))
        pclass.append(int(row[2]))     
        age_value = row[5]
        fare_value = row[9]

        if age_value != '' and age_value != 'NaN':
            age.append(float(age_value))
        else:
            age.append(None)
            
        if fare_value != '' and fare_value != 'NaN':
            fare.append(float(fare_value))
        else:
            fare.append(None)
    except (ValueError, IndexError):
        continue  
def calculate_correlation(x, y):
    filtered_data = [(xi, yi) for xi, yi in zip(x, y) if xi is not None and yi is not None]
    n = len(filtered_data)
    
    if n == 0:
        return None
    
    mean_x = sum(xi for xi, _ in filtered_data) / n
    mean_y = sum(yi for _, yi in filtered_data) / n
    
    numerator = sum((xi - mean_x) * (yi - mean_y) for xi, yi in filtered_data)
    denominator_x = sum((xi - mean_x) ** 2 for xi, _ in filtered_data)
    denominator_y = sum((yi - mean_y) ** 2 for _, yi in filtered_data)

    correlation = numerator / (denominator_x ** 0.5 * denominator_y ** 0.5) if denominator_x and denominator_y else None
    return correlation

age_fare_corr = calculate_correlation(age, fare)
print(f"Correlation between Age and Fare: {age_fare_corr:.2f}")
correlations = {
    'Survived_Age': calculate_correlation(survived, age),
    'Survived_Pclass': calculate_correlation(survived, pclass),
}

for key, value in correlations.items():
    print(f"Correlation between {key.replace('_', ' and ')}: {value:.2f}" if value is not None else f"{key.replace('_', ' and ')}: Not computable")

plt.figure(figsize=(10, 8))

plt.subplot(2, 1, 1)
plt.scatter(age, survived, alpha=0.5)
plt.title('Scatter plot of Age and Survival')
plt.xlabel('Age')
plt.ylabel('Survived')

filtered_fare = [f for f, s in zip(fare, survived) if f is not None and s is not None]
filtered_survived = [s for f, s in zip(fare, survived) if f is not None and s is not None]

plt.subplot(2, 1, 2)
plt.scatter(filtered_fare, filtered_survived, alpha=0.5)
plt.title('Scatter plot of Fare and Survival')
plt.xlabel('Fare')
plt.ylabel('Survived')

plt.tight_layout()
plt.show()
```
